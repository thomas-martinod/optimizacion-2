### Problem 1
Plotting the function $f(x) = f(x_1, x_2) = x_1 e^{-{x_1}^2 - {x_2}^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f
def f(x1, x2):
    return x1 * np.exp(-(x1**2 + x2**2))

# Create a grid of x1 and x2 values
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X1, X2, Z, cmap='viridis')

# Set labels
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1, x_2)$', labelpad=10)  # Adjust labelpad to change the distance of the Z label

# Set title in LaTeX format
plt.title('$f(x)$', fontsize=14)

# Show the plot
plt.show()
```

### Problem 2
Plotting the function $f(x, y) = (x-2)^2 + (y-2)^2$ with the constrains $x+y \geq 4$ and $2x-y \geq 0$.

```python
# Define the function f(x, y)
def f(x, y):
    return (x - 2)**2 + (y - 2)**2

# Create a grid of x and y values
x = np.linspace(0, 4, 400)
y = np.linspace(0, 4, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Define the constraint functions
constraint1 = x + y - 4
constraint2 = 2*x - y

# Create a contour plot for the function
plt.contour(X, Y, Z, levels=np.linspace(0, 8, 20), cmap='viridis')

# Fill the region satisfying the constraints
plt.fill_between(x, x-4, color='gray', alpha=0.5, label='Regin Factible')
plt.fill_between(x, 2*x, color='gray', alpha=0.5)

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.title(r'2D Proyection of $f(x, y) = (x-2)^2 + (y-2)^2$')
plt.grid(True)
plt.show()
```
